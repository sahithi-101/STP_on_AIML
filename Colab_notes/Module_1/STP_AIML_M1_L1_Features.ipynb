## What is Machine Learning?

Machine learning is a branch of artificial intelligence that enables computers to learn patterns from data without being explicitly programmed. Instead of writing specific rules, we train algorithms on examples to make predictions or decisions.

### Types of Machine Learning

**1. Supervised Learning**
- The algorithm learns from labeled data (input-output pairs)
- Goal: Learn a mapping function from inputs to outputs
- Two main types:
  - **Classification**: Predicting categories (e.g., spam/not spam, digit 0-9)
  - **Regression**: Predicting continuous values (e.g., house prices, temperature)

**2. Unsupervised Learning**
- The algorithm finds patterns in unlabeled data
- Goal: Discover hidden structure or relationships
- Examples: Clustering, dimensionality reduction

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the iris dataset (flower classification)
iris = load_iris()
X, y = iris.data, iris.target

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a classifier
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)

# Make predictions
predictions = classifier.predict(X_test)

# Evaluate
accuracy = accuracy_score(y_test, predictions)
print(f"Classification Accuracy: {accuracy:.2%}")
print(f"Predicted classes: {predictions[:5]}")
print(f"Actual classes: {y_test[:5]}")
````

- Features might include: square footage, number of bedrooms, location, age of house

### Why CSV Format?

CSV (Comma-Separated Values) is one of the most common formats for ML datasets because:

1. **Structured**: Data is organized in rows and columns (tabular format)
2. **Easy to read**: Both humans and machines can parse it easily
3. **Universal**: Works with all ML libraries (scikit-learn, pandas, TensorFlow, etc.)

**CSV Structure:**
```
feature1,feature2,feature3,target
1.2,3.4,5.6,0
2.3,4.5,6.7,1
3.4,5.6,7.8,0
```

Each row = one sample/observation
Each column = one feature
Last column = target/label (what we want to predict)

# Part 1: Features of text
How do we apply machine learning on text? We can't directly use the text as input to our algorithms. We need to convert them to features. In this notebook, we will explore a simple way of converting text to features.

Let us download a few documents off Wikipedia.

````python
import wikipedia

# Fix for Wikipedia PageError
topic1 = 'Giraffe'
topic2 = 'Elephant'
topic3 = 'Rose'

def get_wikipedia_content(topic, lang='en'):
    """Safely fetch Wikipedia content with error handling"""
    wikipedia.set_lang(lang)
    try:
        # Disable auto-suggest to prevent title mismatches
        page = wikipedia.page(topic, auto_suggest=False)
        return page.content
    except wikipedia.exceptions.DisambiguationError as e:
        # If multiple pages exist, use the first option
        print(f"Disambiguation for '{topic}': using '{e.options[0]}'")
        return wikipedia.page(e.options[0]).content
    except wikipedia.exceptions.PageError:
        # If page not found, try with auto-suggest enabled
        print(f"Page not found for '{topic}', trying auto-suggest...")
        return wikipedia.page(topic).content

# Fetch content
eng1 = get_wikipedia_content(topic1, 'en')
eng2 = get_wikipedia_content(topic2, 'en')
eng3 = get_wikipedia_content(topic3, 'en')
fr1 = get_wikipedia_content(topic1, 'fr')
fr2 = get_wikipedia_content(topic2, 'fr')
fr3 = get_wikipedia_content(topic3, 'fr')

print("✓ Successfully fetched all Wikipedia pages")
print(f"English {topic1} length: {len(eng1)} characters")
print(f"English {topic2} length: {len(eng2)} characters")
print(f"English {topic3} length: {len(eng3)} characters")
print(f"French {topic1} length: {len(fr1)} characters")
print(f"French {topic2} length: {len(fr2)} characters")
print(f"French {topic3} length: {len(fr3)} characters")
````

We need to clean this up a bit. Let us remove all the special characters and keep only 26 letters and space. Note that this will remove accented characters in French also. We are also removing all the numbers and spaces. So this is not an ideal solution.

````python
def cleanup(text):
  text = text.lower()  # make it lowercase
  text = re.sub('[^a-z]+', '', text) # only keep characters
  return text
````

````python
eng1 = cleanup(eng1)
eng2 = cleanup(eng2)
eng3 = cleanup(eng3)
fr1 = cleanup(fr1)
fr2 = cleanup(fr2)
fr3 = cleanup(fr3)
````

Now let us calculate the frequency of the character n-grams. N-grams are groups of characters of size n. A unigram is a single character and a bigram is a group of two characters and so on.

Let us count the frequency of each character in a text and plot it in a histogram.

````python
# convert a tuple of characters to a string
def tuple2string(tup):
  st = ''
  for ii in tup:
    st = st + ii
  return st

# convert a tuple of tuples to a list of strings
def key2string(keys):
  return [tuple2string(i) for i in keys]

# plot the histogram
def plothistogram(ngram):
  keys = key2string(ngram.keys())
  values = list(ngram.values())

  # sort the keys in alphabetic order
  combined = zip(keys, values)
  zipped_sorted = sorted(combined, key=lambda x: x[0])
  keys, values = map(list, zip(*zipped_sorted))
  plt.bar(keys, values)
````

Another way to visualize bigrams is to use a 2-dimensional graph.

````python
def plotbihistogram(ngram):
  freq = np.zeros((26,26))
  for ii in range(26):
    for jj in range(26):
      freq[ii,jj] = ngram[(chr(ord('a')+ii), chr(ord('a')+jj))]
  plt.imshow(freq, cmap = 'jet')
  return freq
````

Let us look at the top 10 ngrams for each text.

````python
from IPython.core.debugger import set_trace

def ind2tup(ind):
  ind = int(ind)
  i = int(ind/26)
  j = int(ind%26)
  return (chr(ord('a')+i), chr(ord('a')+j))

def ShowTopN(bifreq, n=10):
  f = bifreq.flatten()
  arg = np.argsort(-f)
  for ii in range(n):
    print(f'{ind2tup(arg[ii])} : {f[arg[ii]]}')
````

We observe that the bigrams are similar across different topics but different across languages. Thus, the bigram frequency is a good feature for distinguishing languages, but not for distinguishing topics.

Thus, we were able to convert a many-dimensional input (the text) to 26 dimesions (unigrams) or 26*26 dimensions (bigrams).

````python
print("EXPERIMENT 1: COMPARING MULTIPLE LANGUAGES")

# Fetch documents in multiple languages
languages_to_test = {
    'English': 'en',
    'French': 'fr',
    'German': 'de',
    'Spanish': 'es',
    'Italian': 'it'
}

topic = 'Computer'  # Using same topic across languages

language_texts = {}

for lang_name, lang_code in languages_to_test.items():
    try:
        text = get_wikipedia_content(topic, lang_code)
        language_texts[lang_name] = cleanup(text)
        print(f"Fetched {lang_name}: {len(text)} characters")
    except Exception as e:
        print(f"Failed to fetch {lang_name}: {e}")


# Analyze bigrams for each language
print("\n TOP 10 BIGRAMS PER LANGUAGE:")

language_bigrams = {}

for lang_name, text in language_texts.items():
    bigrams = Counter(ngrams(text, 2))
    language_bigrams[lang_name] = bigrams
    
    print(f"\n{lang_name}:")
    print("-"*70)
    for bigram, count in bigrams.most_common(10):
        print(f"  {tuple2string(bigram):>5s} : {count:5d}")

# Visualize bigram distributions
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

for idx, (lang_name, bigrams) in enumerate(language_bigrams.items()):
    if idx < 6:
        freq = plotbihistogram(bigrams)
        im = axes[idx].imshow(freq, cmap='jet')
        axes[idx].set_title(f'{lang_name} Bigram Distribution', 
                           fontsize=12, fontweight='bold')
        axes[idx].set_xlabel('Second Character')
        axes[idx].set_ylabel('First Character')
        plt.colorbar(im, ax=axes[idx])

# Remove extra subplot if odd number of languages
if len(language_bigrams) < 6:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()

# Compare characteristic bigrams across languages
print("DISTINGUISHING FEATURES BY LANGUAGE")

for lang_name, bigrams in language_bigrams.items():
    # Find unique or highly frequent bigrams
    total = sum(bigrams.values())
    distinctive = []
    
    for bigram, count in bigrams.most_common(50):
        percentage = (count / total) * 100
        if percentage > 1.0:  # Bigrams that appear more than 1% of the time
            distinctive.append((tuple2string(bigram), percentage))
    
    print(f"\n{lang_name} distinctive patterns (>1% frequency):")
    for bigram, pct in distinctive[:5]:
        print(f"  '{bigram}': {pct:.2f}%")
````

````python
print("EXPERIMENT 2: DIFFERENT TOPICS & SOURCES")

# Compare very different topics
diverse_topics = {
    'Science': 'Quantum_Physics',
    'Sports': 'Football',
    'Food': 'Pizza',
    'History': 'World_War_II',
    'Nature': 'Ocean'
}

topic_texts_en = {}

for category, topic in diverse_topics.items():
    try:
        text = get_wikipedia_content(topic, 'en')
        topic_texts_en[category] = cleanup(text)
        print(f"✓ {category} ({topic}): {len(text)} characters")
    except Exception as e:
        print(f"✗ Failed {category}: {e}")

# Analyze bigrams for different topics
topic_bigrams = {}

for category, text in topic_texts_en.items():
    bigrams = Counter(ngrams(text, 2))
    topic_bigrams[category] = bigrams

# Compare bigram distributions across topics
print("\n BIGRAM COMPARISON ACROSS TOPICS:")

# Calculate similarity between topics
def bigram_similarity(bigrams1, bigrams2, top_n=100):
    """
    Calculate similarity between two bigram distributions
    using overlap of top N bigrams
    """
    top1 = set([bg for bg, _ in bigrams1.most_common(top_n)])
    top2 = set([bg for bg, _ in bigrams2.most_common(top_n)])
    
    intersection = len(top1 & top2)
    union = len(top1 | top2)
    
    return intersection / union if union > 0 else 0

# Create similarity matrix
categories = list(topic_bigrams.keys())
similarity_matrix = np.zeros((len(categories), len(categories)))

for i, cat1 in enumerate(categories):
    for j, cat2 in enumerate(categories):
        similarity_matrix[i, j] = bigram_similarity(
            topic_bigrams[cat1], 
            topic_bigrams[cat2]
        )

# Visualize similarity
plt.figure(figsize=(10, 8))
im = plt.imshow(similarity_matrix, cmap='YlOrRd', aspect='auto')
plt.colorbar(im, label='Similarity Score')
plt.xticks(range(len(categories)), categories, rotation=45, ha='right')
plt.yticks(range(len(categories)), categories)
plt.title('Topic Similarity Based on Bigrams', fontsize=14, fontweight='bold')

# Add values to cells
for i in range(len(categories)):
    for j in range(len(categories)):
        text = plt.text(j, i, f'{similarity_matrix[i, j]:.2f}',
                       ha="center", va="center", color="black", fontsize=10)

plt.tight_layout()
plt.show()

print("\n Key Finding:")
print("   Different topics in the same language have very similar bigram")
print("   distributions, confirming bigrams are language-specific features!")

# Show top bigrams for each topic
print("TOP 5 BIGRAMS PER TOPIC:")

for category, bigrams in topic_bigrams.items():
    print(f"\n{category}:")
    for bigram, count in bigrams.most_common(5):
        print(f"  {tuple2string(bigram):>5s} : {count:5d}")
````

````python
print("EXPERIMENT 3: TRIGRAMS AND HIGHER N-GRAMS")

# Use English and French texts for comparison
sample_texts = {
    'English': eng1,
    'French': fr1
}

# Calculate n-grams from 1 to 5
ngram_range = range(1, 6)

for lang_name, text in sample_texts.items():
    print(f"\n{'='*70}")
    print(f"{lang_name.upper()} N-GRAM ANALYSIS")
    print(f"{'='*70}")
    
    for n in ngram_range:
        ngrams_n = Counter(ngrams(text, n))
        
        print(f"\n{n}-grams (Top 10):")
        print("-"*70)
        
        for gram, count in ngrams_n.most_common(10):
            gram_str = tuple2string(gram)
            print(f"  '{gram_str:10s}' : {count:5d}")

# Visualize trigram frequencies
print("VISUALIZING TRIGRAM DISTRIBUTIONS")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for idx, (lang_name, text) in enumerate(sample_texts.items()):
    trigrams = Counter(ngrams(text, 3))
    
    # Get top 30 trigrams
    top_trigrams = trigrams.most_common(30)
    labels = [tuple2string(tg[0]) for tg in top_trigrams]
    counts = [tg[1] for tg in top_trigrams]
    
    axes[idx].barh(range(len(labels)), counts, color='steelblue', alpha=0.7)
    axes[idx].set_yticks(range(len(labels)))
    axes[idx].set_yticklabels(labels, fontsize=8)
    axes[idx].set_xlabel('Frequency', fontsize=12)
    axes[idx].set_title(f'{lang_name} - Top 30 Trigrams', 
                        fontsize=14, fontweight='bold')
    axes[idx].invert_yaxis()
    axes[idx].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# Compare n-gram diversity across languages
print("N-GRAM DIVERSITY ANALYSIS")

diversity_data = []

for n in ngram_range:
    for lang_name, text in sample_texts.items():
        ngrams_n = Counter(ngrams(text, n))
        
        unique_count = len(ngrams_n)
        total_count = sum(ngrams_n.values())
        diversity = unique_count / total_count if total_count > 0 else 0
        
        diversity_data.append({
            'N': n,
            'Language': lang_name,
            'Unique': unique_count,
            'Total': total_count,
            'Diversity': diversity
        })
        
        print(f"{lang_name:10s} {n}-grams: {unique_count:6d} unique / "
              f"{total_count:7d} total (diversity: {diversity:.4f})")

# Visualize diversity trends
df_diversity = pd.DataFrame(diversity_data)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1: Unique n-grams by n
for lang_name in sample_texts.keys():
    lang_data = df_diversity[df_diversity['Language'] == lang_name]
    ax1.plot(lang_data['N'], lang_data['Unique'], 
             marker='o', linewidth=2, markersize=8, label=lang_name)

ax1.set_xlabel('N-gram Size', fontsize=12)
ax1.set_ylabel('Number of Unique N-grams', fontsize=12)
ax1.set_title('Unique N-grams vs N-gram Size', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xticks(ngram_range)

# Plot 2: Diversity by n
for lang_name in sample_texts.keys():
    lang_data = df_diversity[df_diversity['Language'] == lang_name]
    ax2.plot(lang_data['N'], lang_data['Diversity'], 
             marker='s', linewidth=2, markersize=8, label=lang_name)

ax2.set_xlabel('N-gram Size', fontsize=12)
ax2.set_ylabel('Diversity (Unique/Total)', fontsize=12)
ax2.set_title('N-gram Diversity vs N-gram Size', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xticks(ngram_range)

plt.tight_layout()
plt.show()
````

Let us look at another, more complicated feature. We will count the number black pixels that are surrounded on four sides by non-black pixels, or "hole pixels".

````python
def cumArray(img):
  img2 = img.copy()
  for ii in range(1, img2.shape[1]):
    img2[ii,:] = img2[ii,:] + img2[ii-1,:]  # for every row, add up all the rows above it.
  img2 = img2>0
  return img2

def getHolePixels(img):
  im1 = cumArray(img)
  im2 = np.rot90(cumArray(np.rot90(img)), 3) # rotate and cumulate it again for differnt direction
  im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
  im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
  hull =  im1 & im2 & im3 & im4 # this will create a binary image with all the holes filled in.
  hole = hull & ~ (img>0) # remove the original digit to leave behind the holes
  return hole
````

Visualize a few:

````python
hole2 = np.array([getHolePixels(i).sum() for i in no2])
hole1 = np.array([getHolePixels(i).sum() for i in no1])
hole0 = np.array([getHolePixels(i).sum() for i in no0])

plt.hist(hole2, alpha=0.7);
plt.hist(hole1, alpha=0.7);
plt.hist(hole0, alpha=0.7);
````

This feature works even better to distinguish between one and zero.


Now let us try the number of pixels in the 'hull' or the number with the holes filled in:

````python
def getHullPixels(img):
  im1 = cumArray(img)
  im2 = np.rot90(cumArray(np.rot90(img)), 3) # rotate and cumulate it again for differnt direction
  im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
  im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
  hull =  im1 & im2 & im3 & im4 # this will create a binary image with all the holes filled in.
  return hull
````

````python
imgs = [no2[456,:,:],no1[456,:,:],  no0[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getHullPixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
````

Let us try one more feature, where we look at the number of boundary pixels in each image.

````python
def minus(a, b):
  return a & ~ b

def getBoundaryPixels(img):
  img = img.copy()>0  # binarize the image
  rshift = np.roll(img, 1, 1)
  lshift = np.roll(img, -1 ,1)
  ushift = np.roll(img, -1, 0)
  dshift = np.roll(img, 1, 0)
  boundary = minus(img, rshift) | minus(img, lshift) | minus(img, ushift) | minus(img, dshift)
  return boundary
````

````python
imgs = [no2[456,:,:], no1[456,:,:],  no0[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getBoundaryPixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
````

````python
bound2 = np.array([getBoundaryPixels(i).sum() for i in no2])
bound1 = np.array([getBoundaryPixels(i).sum() for i in no1])
bound0= np.array([getBoundaryPixels(i).sum() for i in no0])

plt.hist(bound2, alpha=0.7);
plt.hist(bound1, alpha=0.7);
plt.hist(bound0, alpha=0.7);
````
